### Set Labels of Nodes

```python
labels=['1 - F7',
'2 - F3',
'3 - F4',
'4 - F8',
'5 - T3',
'6 - C3',
'7 - Cz',
'8 - C4',
'9 - T4',
'10 - T5',
'11 - P3',
'12 - Pz',
'13 - P4',
'14 - T6',
'15 - O1',
'16 - O2']
```

### Set Limited-Bands

```python
band={'delta':0 , 'theta':1 , 'alpha' : 2 , 'beta' : 3, 'gamma' : 4}
```

## **Set Parameters**

```python
matlab_file='subject_normal_1.mat'
band_sel='alpha'
title={'class_name' : 'normal',
       'subject_num' : 1}
```

### Connectivity Matrix

```python
X = scipy.io.loadmat(matlab_file)
X= X['connectivity_matrix']
plotting.plot_matrix(X[:,:,band[band_sel]], labels=labels, colorbar=True)
plt.title('Class: {} - Subject: {}th - Band: {}'.format(title['class_name'], title['subject_num'], band_sel))
```

### Graph Theory

```python
plotting.plot_connectome(X[:,:,band[band_sel]], coords,
                         edge_threshold="80%", colorbar=True)

plotting.show()
```

### 3D View Graph

```python
view = plotting.view_connectome(X[:,:,band[band_sel]], coords, edge_threshold='80%')
view
```
